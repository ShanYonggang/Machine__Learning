```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# 读取数据
data = pd.read_csv(r"F:\0_个人学习\3_数据分析与挖掘\4_机器学习\2_线性回归\job.csv")
x = np.array(data[:]['Level'])
y = np.array(data[:]['Salary'])
```

```python
# 绘制散点图
plt.scatter(x,y)
plt.show()
```

```python
y.reshape(-1,1)
```

```python
# 使用一元线性回归
model_1 = LinearRegression()
x = x.reshape(-1,1)
y = y.reshape(-1,1)
plt.scatter(x,y)
model_1.fit(x,y)
plt.plot(x,model_1.predict(x),'r')
plt.show()
```

```python
# 定义多项式回归,degree的值可以调节多项式的特征
for i in range(7):
    # 主要更改degree的值来观察拟合曲线的变化
    poly_model = PolynomialFeatures(degree=i)
    # 特征值处理
    x_data = poly_model.fit_transform(x)
    # 定义回归模型
    model_2 = LinearRegression()
    # 训练模型
    model_2.fit(x_data,y)
    # 绘图
    plt.plot(x, y, 'b.')
    plt.plot(x,model_2.predict(x_data),'g')
    plt.title('Truth or Bluff (Polynomial Regression)')
    plt.xlabel('Position level')
    plt.ylabel('Salary')
    plt.show()
```

```python
# 以下选取degree值为5时候使得曲线更为平滑
poly_model = PolynomialFeatures(degree=5)
# 特征值处理
x_data = poly_model.fit_transform(x)
# 定义回归模型
model_2 = LinearRegression()
# 训练模型
model_2.fit(x_data,y)
# 画图
plt.plot(x, y, 'b.')
x_test = np.linspace(1,10,100)
x_test = x_test[:,np.newaxis]
plt.plot(x_test, model_2.predict(poly_model.fit_transform(x_test)), c='r')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
# 输入数据
x = np.array([
            [1,3,3],
            [1,4,3],
            [1,1,1],
            [1,0,2]
            ])
# 输出数据
y = np.array([
            [1],
            [1],
            [-1],
            [-1]
            ])
# 权值初始化
w = (np.random.random([3,1])-0.5)*2
# 转化成矩阵
x_mat = np.mat(x)
y_mat = np.mat(y)
w_mat = np.mat(w)
# 学习率
lr = 0.11
# 神经网络输出
O = 0
def update():
    global x_mat,y_mat,w_mat,lr
    O = np.sign(x_mat*w_mat)
    dela_w = lr*x_mat.T*(y_mat-O)/int(x_mat.shape[0])
    w_mat = w_mat + dela_w
```

```python
for i in range(100):
    update()
    print(w_mat)
    print(i)
    O = np.sign(x_mat*w_mat)
    print(O)
    if (O==np.mat(y)).all():
        print('finished!')
        print('循环次数为{}'.format(i+1))
        break;
#正样本
x1 = [3,4]
y1 = [3,3]
#负样本
x2 = [1,0]
y2 = [1,2]

#计算分界线的斜率以及截距
k = -w_mat[1]/w_mat[2]
d = -w_mat[0]/w_mat[2]
k = np.array(k)[0]
d = np.array(d)[0]
print('k=',k)
print('d=',d)

xdata = (0,5)
plt.figure()
plt.plot(xdata,xdata*k+d,'r')
plt.scatter(x1,y1,c='b')
plt.scatter(x2,y2,c='y')
plt.show()
```
